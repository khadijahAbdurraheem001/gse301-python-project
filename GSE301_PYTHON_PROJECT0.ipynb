```python
#2.1:Conditional statements for grading
def get_grade(score:float)->str:
  if score>=70 and score<=100:
    return "A"
  elif score>=60:
    return "B"
  elif score>=50:
    return "C"
  elif score>=45:
    return "D"
  elif score>=40:
    return "E"
  else:
    return "F"
def grade_feedback(grade):
  match grade:
    case "A":
      print("excellent")
    case "B":
      print("very good")
    case "C":
      print("good")
    case "D":
      print("fair")
    case "E":
      print("poor results")
    case "F":
      print("failed")

score=int(input("enter your score:(0-100)"))
grade=get_grade(score)
print("grade:", grade)
grade_feedback(grade)

#Type validation and conversion
age_input = input("enter your age:")
cgpa_input = input("enter your cgpa:")
age = int(age_input)
cgpa = float(cgpa_input)

if age < 16 or age > 40:
  print("invalid age:age must be between 16 and 40")
elif cgpa <0.00 or cgpa >5.00:
  print("invalid cgpa:cgpa must be between 0.0 and 5.0")
else:
  print("age and cgpa are valid")
  print("age:", age)
  print("cgpa:", cgpa)
```
